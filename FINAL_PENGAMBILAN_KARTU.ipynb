```python
import random
```

```python
fullArr = [i for i in range(1,51)]
cardsTaken = []
totalSum = []
whichCardsTaken = []
```

```python
def takeCard(x):
  sum=0
  cardsTaken.append(x)
  dummyList = []
  while(x>0):
    forSum = random.randint(0,len(fullArr)-1)
    dummyList.append(fullArr[forSum])
    sum+=fullArr.pop(forSum)
    x-=1
  whichCardsTaken.append(list(dummyList))
  totalSum.append(sum)
```

```python
for _ in range(100000): # angka 1000 bisa diganti dengan banyak sample yang kita mau
  x = random.randint(1,50)
  takeCard(x)
  fullArr = [i for i in range(1,51)]
```

```python
LwhichCardsTaken = list(whichCardsTaken)
print(cardsTaken)
print(totalSum)
print(LwhichCardsTaken)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
```

Distribusi normal pengambilan kartu secara acak (MASIH PERLU DIREVISI)


```python
cardsTakenArr = np.array(cardsTaken)
xCards = np.sort(cardsTakenArr)
cardsMean = statistics.mean(xCards)
sdCards = statistics.stdev(xCards)

plt.plot(xCards, norm.pdf(xCards, cardsMean, sdCards))
plt.xticks([i for i in range(0,55,5)])
plt.show()
```

Distribusi normal jumlah kartu secara acak (MASIH PERLU DIREVISI)

```python
totalSumArr = np.array(totalSum)
xSum = np.sort(totalSumArr)
sumMean = statistics.mean(xSum)
sdSum = statistics.stdev(xSum)

plt.plot(xSum,norm.pdf(xSum,sumMean,sdSum))
plt.xticks([i for i in range(0,1400,100)])
plt.show()
```

```python
colors = (0,0,0)
area = np.pi*3

# Plot
plt.scatter(cardsTaken, totalSum)
plt.title('Pengambilan Kartu')
plt.xlabel('Banyak Kartu terambil')
plt.ylabel('Jumlah Kartu terambil')
plt.xticks([i for i in range(0,55,5)])
plt.yticks([i for i in range(0,1600,100)])
plt.show()
```
